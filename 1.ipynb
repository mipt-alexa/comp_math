```python
# Welcome to Jupyter!

a = [-512,  2304, -4608, 5376,  -4032,  2016, -672, 144, - 18, 1]
len(a)
```

```python
def horner_method(x, *coef):
    n = len(coef)-1
    p = coef[n]
    for i in range(n-1, -1, -1):
        p = x*p + coef[i]
    
    return p

horner_method(2, *a)
```

```python
def func(x):
    return (x-2)**9
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(1.92, 2.08, 1600)

horner = np.vectorize(horner_method)
f = np.vectorize(func)

plt.plot(X, horner(X, *a))
plt.plot(X, f(X), color='red')

plt.show()
```
