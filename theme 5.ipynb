#7.1б

$f = \dfrac{ x^3}{3} + x^2$

Сначала алгебраический метод:

$f' = x^2 + 2x = 0 $

$f'' = 2x + 2$


$x = -2 $ соответствует минимуму

Дихотомия с точностью $\delta = 10^{-4}$, $\varepsilon = 10^{-5}$ локализуем минимум на отрезке [-3,3]

```python
def f(x):
    return 1/3* np.power(x,3) + x*x
```

```python
a = -3
b = 3

d = 1e-4
e = 1e-5

while b-a > 2*d:
    c = (a+b)/2

    if f(c - e) < f(c + e):
        b = c + e
    else:
        a = c - e

print((a+b)/2)
```

локализуем один минимум в области [0,1]x[0,1], второй [-0.5, 0.5]x[0,-1]

```python
def dichotomy_2d(x1, x2, y1, y2): 
    a = np.array([x1,y1])
    b = np.array([x2,y2])
    c = np.empty(2)

    d = 1e-4
    e = 1e-5

    while b[0]-a[0] > 2*d and b[1]-a[1] > 2*d:
        c[0] = (a[0]+b[0])/2
        c[1] = (a[1]+b[1])/2

        if f(c[0] - e, c[1]) < f(c[0] + e, c[1]):
            b[0] = c[0] + e
        else:
            a[0] = c[0] - e

        c[0] = (a[0]+b[0])/2

        if f(c[0], c[1] - e) < f(c[0], c[1] + e):
            b[1] = c[1] + e
        else:
            a[1] = c[1] - e


    print("x ", (a[0]+b[0])/2)
    print("y ", (a[1]+b[1])/2)

dichotomy_2d(0., 1., 0., 1.)

dichotomy_2d(-0.5, 0.5, -1., 0.)
```
