## 12.4в

$(0.5)^x + 1 = (x – 1)^2$




$f(x) = (0.5)^x + 1 - (x – 1)^2$

$f' = 0.5^x \ln (0.5) - 2 (x-1)$


```python
x = np.linspace(-6,6,100)

plt.figure(figsize=(10, 5), dpi=80)

plt.plot(x, np.power(0.5, x) + 1 - np.power((x-1),2), label="f")
# plt.plot(x, np.power(0.5, x)*np.log(0.5)- 2*(x-1), label="df/dx")
plt.legend()
plt.grid()
```

Локализуем корни: 

$x_1 \in [-6,-4]$

$x_2 \in [-2,0]$

$x_3 \in [2,4]$


Попробуем итерационный процесс 

$x = 1 + \sqrt{0.5^x +1 }$

$\phi(x) > 0 \;\; \forall \, x$,  

 $|\phi(x)'| < 1 \;\;\; \forall \;\; x >0$ 

Сходится к $x_3$ даже для любого начального приближения (потому что при $x_0 < 0$ на первом шаге $x^{(1)} > 0$)

```python
def proc(x0):
    x = np.array([100.,0.])

    x[1] = x0

    while abs(x[0]-x[1]) > 1e-4:
        x[0] = x[1]
        x[1] = 1 + np.sqrt(np.power(0.5, x[1]) +1)

    print(x[1])

proc(-10)
proc(-2)
proc(0)
proc(1)
proc(3)
proc(-1000)
```

2 пары решений, симметричных от-но нуля

```python

x0 = 0.1
y0 = 0.3

x = np.array([x0,y0])
norm = 0

while abs(np.linalg.norm(x, 2) - norm) > 1e-3:
    norm = np.linalg.norm(x, 2)

    x[0] = np.sqrt(np.tan(x[0]*x[1] + 0.4 ))
    x[1] = np.sqrt(0.5*1 - 0.5*0.6*x[0]*x[0])

print(x)
```
